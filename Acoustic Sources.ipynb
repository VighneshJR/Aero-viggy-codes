# ACOUSTIC SOURCES

The wave equation in general is given by $$\nabla ^2 \psi =-\frac{k^2}{\omega ^2} \frac{\partial ^2 \psi}{\partial t^2}$$. In polar coordinates that become 
$$ \frac{1}{r^2} \frac{\partial}{\partial r} (r^2\frac{\partial \psi}{\partial r})  = -\frac{k^2}{\omega ^2} \frac{\partial ^2 \psi}{\partial t^2}$$ <br>
Since our point source is isotropic , we have the $\theta$ and $\phi$ dependance to be nullified. So after removing theta terms and lets attempt a solution $\psi = f(r)sin(kr - \omega t)$ . This turns out to be our monopole equation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


def monopole(t):
    x,y =np.mgrid[-6:6:100j,-5:5:100j]
    r1=np.sqrt(x**2 + y*y)
    psi=20*np.sin(np.pi*r1 - 5*t)/r1
    plt.contourf(x,y,psi,np.linspace(-25,60))
    plt.axis("equal")
    plt.colorbar()
    plt.xlabel("x axis")
    plt.ylabel("y axis")

def dipole(t):
    x,y =np.mgrid[-6:6:70j,-5:5:80j]
    r1=np.sqrt((x-0.1)**2 + y*y)
    r2=np.sqrt((x+0.1)**2 + y*y)
    psi=200*np.sin(np.pi*r1 - t)/r1 - 200*np.sin(np.pi*r2 - t)/r2
    plt.contourf(x,y,psi,np.linspace(-200,200),cmap='plasma')
    plt.axis("equal")
    plt.colorbar()
    plt.xlabel("x axis")
    plt.ylabel("y axis")

def quadrapole(t,s=5):
    x,y =s*np.mgrid[-1.2:1.2:100j,-1:1:100j]
    r1=np.sqrt((x-0.1)**2 + y*y)
    r2=np.sqrt((x+0.1)**2 + y*y)
    r3=np.sqrt((y-0.1)**2 + x*x)
    r4=np.sqrt((y+0.1)**2 + x*x)
    #nul=np.ones_like(x)
   # cond = np.logical_or(r1<0.05,r2<0.05,r3<0.05,r4<0.05)
    psi=20*np.sin(np.pi*r1 - t)/r1 + 20*np.sin(np.pi*r2 - t)/r2- 20*np.sin(np.pi*r3 - t)/r3 - 20*np.sin(np.pi*r4 - t)/r4
    #psi[cond]=0
    plt.contourf(x,y,psi,np.linspace(-2.5,2.5,50),cmap='plasma')
    plt.axis("equal")
    plt.colorbar()
    plt.xlabel("x axis")
    plt.ylabel("y axis")
 
```

## **Monopole** 
We have a point source located at origin and the wave function is $$\psi(r,t) = \frac{\mathbf{A}}{r}sin(kr - \omega t)$$ where k is the wave number and $\omega$ is angular frequency. Here it is a point wave source which is isotropic.Lets do a contour plot of its value strictly restricting ourselves to the x-y plane

```python
monopole(0)
plt.title("Monpole source of wavelength 2 m")    
plt.show()   
```

## **Dipole**
For constructing a dipole we put two monopoles differing by phase $\pi$ a small distace apart on the x axis for this problem. We can also notice that x = 0 is a nodal plane for this wave

```python
dipole(0)
plt.title("dipole at t =0 sec ")
plt.show()
```

## **Quadrapole**
Basically 4 monopoles but kept apart by small distances on the axis and but the monopoles on the same axis have same phase while those on different axes have a phase difference of $\pi$. x+y=0 and x-y = 0 are two nodal planes for this system

```python
quadrapole(0)
plt.title("Quadrapole source with wavelenght 2 and at time t = 0  seconds")
plt.show()
```

## General Trivia

If we notice the equation of a monopole , we see that there is an amplitude part and wave part ( sinosoidal ) but we can also consider it as a complex function part. Hence <br> 
$$ \hat{\psi}(r,t) = \frac{A}{r} e^{i(kr - \omega t)} $$
where the amplitude part is $|\psi(r,t)| = \frac{A}{r}$ as we know the point source is isotropic and via conservation of energy , net power  through a spherical surface of any radius around source must be same. Since intensity is proportional to $\psi^{*}\psi$ and power flux is proportional to $|\psi|^2 r^2$ . For r invariance we have $|\psi|$ $\alpha$ $\frac{1}{r}$ , hence the given form of amplitude function.<br>
Lets plot and see the imaginary and real part of the wave function at time t = 0

```python
r= np.linspace(0.01,10,100)
rd =np.linspace(0.25,10,100)
psi = 20*(np.e**(1j*4*r))/r
reals = np.real(psi)
imag = np.imag(psi)
plt.plot(r,reals)
plt.plot(r,20/r,'--')
plt.title("Real part")
plt.legend(["wave function","amplitude"])
plt.xlabel("distance form source")
plt.grid()
```

```python
plt.plot(r,imag)
plt.plot(rd,20/rd,'--')
plt.legend(["wave function","amplitude"])
plt.title("imaginary part")
plt.xlabel("distance form source")
plt.grid()
```
