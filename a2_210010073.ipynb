### Part A

#### Question 1 

We have a uniform flow with velocity aligned at an angle $\alpha$ to the x axis and let the magnitude of velocity be v. Then we have the velocity field $\vec{V}(x,y)$ = v$cos(\alpha)$ $\hat{x}$ + v$sin(\alpha)$ $\hat{y}$. Now we have to find the cartesian expressions for flow poetentials and stream functions. We Know that $$\vec {\nabla \phi} = \vec{V}(x,y)$$ where $\phi(x,y)$ is the potential function. From integration we have $$\phi(x,y)=(vcos\alpha)x + (vsin\alpha)y$$ We have this expression because v and $\alpha$ are uniform across space ie constant wrt x and y. Now in order to find the stream function $\psi(x,y)$ we have to solve the following PDE. $$\frac{\partial \psi}{\partial y}\hat{x} -\frac{\partial \psi}{\partial x}\hat{y} = \vec{V}(x,y)$$ Again after solving this we obtain $$\psi(x,y) = (vcos\alpha)y - (vsin\alpha)x$$ Now we need the complex potential function $\Phi(z)$ where $z=x+iy$ , $i=\sqrt{-1}$. From definiton we have $$\Phi(z) = \phi(x,y) + i\psi(x,y)$$ By plugging in the above expressions we have 
$$\Phi(z) = (vcos\alpha)x + (vsin\alpha)y + i((vcos\alpha)y - (vsin\alpha)x )$$
$$\Phi(z) = [(vcos\alpha) - i(vsin\alpha)]x + [(vsin\alpha) + i(vcos\alpha)]y$$
$$\Phi(z) = [(vcos\alpha) - i(vsin\alpha)]x + [\frac{(vsin\alpha)}{i} + (vcos\alpha)]iy$$
We know that $\frac{1}{i} = -i$ so 
$$\Phi(z) = [(vcos\alpha) - i(vsin\alpha)]x + [ (vcos\alpha)-i(vsin\alpha) ]iy$$
$$\Phi(z) =[(vcos\alpha) - i(vsin\alpha)](x+iy)$$
Finally we have the complex potential function 
$$\Phi(z) = [(vcos\alpha) - i(vsin\alpha)]z$$ <br>

#### Question 1

Lets start with a source of unit strength hence the flow field will have only radial component of velocity if we take the location of the source as the origin. Hence in cartesian coordinates we have the velocity field as $$\vec{V}(x,y) = \frac{1}{2\pi}\frac{x\hat{i}+y\hat{j}}{x^2 + y^2} = u\hat{i} + v\hat{j}$$ Now we have to find out the complex potential for this flow $\Phi(z)$ where z =x +$i$y ;$i =\sqrt{-1}$ . But we know that the complex velocity field is $$\frac{d\Phi(z)}{dz} = u -iv = \frac{1}{2\pi}\frac{x-iy}{x^2+y^2}$$
$$(x+iy)(x-iy)=x^2+y^2$$ so using that we have
$$\frac{d\Phi(z)}{dz} = \frac{1}{2\pi}\frac{1}{x+iy} = \frac{1}{2\pi}\frac{1}{z}$$
Now by just integratiing we have 
$$\Phi(Z) = \frac{1}{2\pi} \mathcal{ln}(z)$$
We got the complex potential function but how to get the normal potential and stream functions from this?<br>
we know that $\Phi(z) = \phi(x,y) + i\psi(x,y)$ and also the polar form of a complex number $$z = (\sqrt{x^2 + y^2})e^{i * tan^{-1}(\frac{y}{x})}$$ Now plugging this in our complex potential we have 
$$\Phi(z) = \frac{1}{2\pi}ln((\sqrt{x^2 + y^2})e^{i * tan^{-1}(\frac{y}{x})})$$ Which is 
$$\Phi(z) = \frac{1}{2\pi}ln(\sqrt{x^2 + y^2}) + i \frac{ tan^{-1}(\frac{y}{x})}{2\pi}$$
In comparison with the definition of complex potential we have <br> the flow potential to be $\phi(x,y) = \frac{1}{4\pi}ln(x^2 + y^2)$ <br> and the stream function to be $\psi(x,y) = \frac{1}{2\pi}tan^{-1}(\frac{y}{x})$

Now with the above defined expressions we can make python functions for velocity fields , potential fields and stream functions

```python
def vel(x,y): #Velocity field for a source 
    u=(0.5/np.pi)*x/(x**2 + y**2)
    v=(0.5/np.pi)*y/(x**2 + y**2)
    cond = x**2 + y**2 < 0.1   
    u[cond]=0 #singularity avoiding adjustments
    v[cond]=0
    return np.array([u,v])
```

```python
X,Y= np.mgrid[-3:3:20j,-3:3:20j]
U,V=vel(X,Y)
plt.quiver(X,Y,U,V,color='b')
plt.axis("equal")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Velocity field from a singular source located at origin") 
plt.show()
```

```python
def poten(x,y):#potential function
    P=(0.25/np.pi)*np.log(x**2 + y**2)
    cond=x**2 +y**2 < 0.01
    P[cond]=0 #singularity avoiding mechanism
    return P
X,Y= np.mgrid[-3:3:400j,-3:3:400j]
phi = poten(X,Y)

plt.contour(X, Y, phi)

plt.colorbar()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("potential lines of radial flow form singular source at origin")
plt.axis('equal')
plt.show()
```

```python
def streamfun(x,y):
    P=(0.5/np.pi)*np.arctan(y/x)
    #cond=x**2 < 0.0001
    #P[cond]=1 #singularity avoiding mechanism
    return P
X,Y= np.mgrid[-3:3:400j,-3:3:400j]
pst = streamfun(X,Y)

plt.contour(X, Y, pst,)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.colorbar()
plt.title("streamlines of radial flow form singular source at origin")
plt.axis('equal')
plt.show()
```

#### Question 2

For this question lets assume a doublet aligned along x axis in the -ve x direction , ie the source at $(-\frac{l}{2},0)$ and the sink at $(\dfrac{l}{2},0)$ both of some strength q and -q such that $ql = \mu$ and we consider the limit $\lim_{l\to 0}$ case for the doublet. We know that the potential fields can be superimposed to get the doublet field. Since we consider a doublet of unit strength we take $\mu = 1$.So the new velocity field is 
$$\phi_{doublet}(x,y) =\lim_{l\to0} \frac{q}{4\pi}ln((x+\frac{l}{2})^2 +y^2)-\frac{q}{4\pi}ln((x-\frac{l}{2})^2 +y^2)$$
$$\phi(x,y) = \frac{\mu}{2\pi}\frac{x}{x^2+y^2}$$
 here we take $\mu =1 $ and subsequently the velocity field is 
$$\vec{V}(x,y) = \frac{\mu}{2\pi}\frac{y^2 -x^2}{(x^2 +y^2)^2}\hat{i} -\frac{\mu}{2\pi}\frac{2xy}{(x^2 +y^2)^2}\hat{j} $$
The stream function is $$\psi(x,y) =\frac{-\mu}{2\pi}\frac{y}{x^2+y^2} $$
The complex potential function when $\mu =1$ is
$$\Phi(z) = \frac{\mu}{2\pi}\frac{x-iy}{x^2+y^2} = \frac{\mu}{2\pi z}$$ where z =x +iy. Now lets use these expressions to plot the corresponding fields in our usual domain

```python
def vel(x,y): #Velocity field for a source 
    u=(0.5/np.pi)*(y**2 - x**2)/(x**2 + y**2)**2
    v=-(0.5/np.pi)*2*x*y/(x**2 + y**2)**2
   
    cond = x**2 + y**2 < 0.4   
    u[cond]= 0 #singularity avoiding adjustments
    v[cond]= 0
    return np.array([u,v])
X,Y= np.mgrid[-3:3:40j,-3:3:40j] #mesh creation for the usual domain
U,V=vel(X,Y)
plt.quiver(X,Y,U,V,color='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.title("Velocity field from a doublet located at origin oriented along negative x") 
plt.show()
```

#### Question 3

We have a point vortex of unit strength ie the circulation $\Gamma$ = 1 . The velocity field will only have a angular(tangential) component(polar coordinates) and for a vortex situated at origin the velocity field is given by:
$$\vec{V}(x,y)=\frac{\Gamma}{2\pi}\frac{-y\hat{i} + x\hat{j}}{x^2+y^2} = u\hat{i} + v\hat{j}$$
From the definiton of complex potential we have $$\frac{d\Phi(z)}{dz}=u(x,y)-iv(x,y) $$ Where u , v,z and i have their usual meanings. Now plugging the expressions in we have the following differential equation:
$$\frac{d\Phi(z)}{dz}=\frac{\Gamma}{2\pi}\frac{-y - ix}{x^2+y^2}=\frac{\Gamma}{2\pi}\frac{-i(-iy + x)}{x^2+y^2}=\frac{\Gamma}{2\pi}\frac{-i}{x+iy}$$
$$\frac{d\Phi(z)}{dz}=\frac{-i\Gamma}{2\pi z}$$
Now after integrating we have our complex potential function:
$$\Phi(z)=\frac{-i\Gamma}{2\pi}ln(z)$$
This can be written in the cartesian form as :
$$\Phi(z)=\frac{-i\Gamma}{2\pi}ln(\sqrt{x^2+y^2}e^{itan^{-1}(\frac{y}{x})})$$
This is using the polar (eulerian ) form of a complex number.On further simplification we have
$$\Phi(z)=\frac{-i\Gamma}{2\pi}(ln(\sqrt{x^2+y^2})+itan^{-1}(\frac{y}{x}))$$
$$\Phi(z)=\frac{\Gamma}{2\pi}[tan^{-1}(\frac{y}{x})-iln(\sqrt{x^2+y^2})]$$
But we also know that $\Phi(z)=\phi(x,y) + i\psi(x,y)$ Where $\phi$ is potential and $\psi$ is stream function. Then directly from the above equation we have :
$$\phi(x,y) = \frac{\Gamma}{2\pi}tan^{-1}(\frac{y}{x})$$
and 
$$\psi(x,y)=\frac{-\Gamma}{4\pi}ln(x^2+y^2) $$
For plotting reasons and by the question we take $\Gamma$ =1

```python
#the velocity field , we are going to plot this in our usual domain
def vel(x,y): #Velocity field for a source 
    v=(0.5/np.pi)*x/(x**2 + y**2)
    u=-(0.5/np.pi)*y/(x**2 + y**2)
    cond = x**2 + y**2 < 0.3   
    u[cond]=0 #singularity avoiding adjustments
    v[cond]=0
    return np.array([u,v])
X,Y= np.mgrid[-3:3:20j,-3:3:20j]
U,V=vel(X,Y)
plt.quiver(X,Y,U,V,color='darkblue')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.title("Velocity field around a point vortex located at origin") 
plt.show()
```

```python
def potfun(x,y):#potential function
    P=(0.5/np.pi)*np.arctan(y/x)
    #cond=x**2 < 0.0001
    #P[cond]=1 #singularity avoiding mechanism
    return P
X,Y= np.mgrid[-3:3:400j,-3:3:400j]
phy = potfun(X,Y)

plt.contour(X, Y, phy)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.colorbar()
plt.title("iso potential lines of point vortex flow")
plt.axis('equal')
plt.show()
```

##### Question 1
Velocity field:<br>

we know that velocity field solutions can be superimposed to match the boundary conditions. so we have our required

$$\vec{V}(x,y)= (1 + \frac{x+0.5}{2\pi((x+0.5)^2 +y^2)})\hat{i} +\frac{y}{2\pi((x+0.5)^2 +y^2)}\hat{j} $$

```python
def velf(x,y):
    u=1 + (0.5/np.pi)*(x+0.5)/((x+0.5)**2+y**2)
    v=(0.5/np.pi)*(y)/((x+0.5)**2+y**2)
    cond=x**2 + y**2 <0.001
    u[cond]=0
    v[cond]=0
    return np.array([u,v])
X,Y= np.mgrid[-0.8-0.5:0.8-0.5:20j,-0.8:0.8:20j]
U,V=velf(X,Y)
plt.quiver(X,Y,U,V,color='b')
plt.axis("equal")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("velocity field of flow past a point source") 
plt.plot([-0.5],[0],"or")
plt.show()
```

##### **pessure and velocity magnitude contours**

```python
def vel_mag(x,y): #velocity magnitude field
    u= 1+ (0.5/np.pi)*(x+0.5)/((x+0.5)**2 +y**2) - (0.5/np.pi)*(x-0.5)/((x-0.5)**2 +y**2)
    v=(0.5/np.pi)*y/((x+0.5)**2 +y**2) - (0.5/np.pi)*y/((x-0.5)**2 +y**2)
    #cond = np.logical_or((x+0.5)**2 + y**2 < 0.00025,(x-0.5)**2 + y**2 < 0.00025) #singularity avoiding mechanisms
    #u[cond]=0
    #v[cond]=0
    
    return np.sqrt(u**2 +v**2)
```

We know that according to the Bernoullis Equation (Ignore gravity here) is $$P + \frac{1}{2}\rho V^2 =  const$$ for all points in a streamline and since this constant is same for all streamlines since prssure and velocity are identically in free stream we have:
$$P_{\infty} + \frac{1}{2}\rho V_{\infty}^2 = P(x,y) + \frac{1}{2}\rho V(x,y)^2$$
For the given above flow , $V(x,y)^2$ is known. Hence the pressure field is :
$$ P(x,y)=P_{\infty} +\frac{1}{2}\rho (V_{\infty}^2 - V(x,y)^2)  $$

For the sake of this problem lets assume that the free stream pressure is 30 units and density is 10 units

```python
Pinf =30 # these can be edited
den=10# these can be edited
#free stream velocity has magnitude 1 as we have discussed above
def P(x,y):
    return Pinf + 0.5*den*(1 - (vel_mag(x,y))**2)#pressure field
```
