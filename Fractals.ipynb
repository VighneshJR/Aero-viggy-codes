 # <center> MANDELBROTH SET </center>
$$f(z) = z^2 + c$$
It is the set of points {c} in complex plane such that the sequence {$z_n$} such that $z_0 = 0$ and $f(z_n) = z_{n+1} \forall n\geq 0  $ is a bounded sequence. It is a  with microstructures and in the below diagram the white region is a poor approximation for the mandelbroth set due to low resolution. The below plot is basically a contour plot of $g(c) = \lim_{n \to \infty}  log(|z_n(c) - z_{n-1}(c)|+\delta)$ and in order to prevent overflow we have taken n as 11. For boundedness we apply the caushy criterion, hence g(c) must be of negative value.
