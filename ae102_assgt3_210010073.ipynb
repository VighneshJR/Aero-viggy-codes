##### part(ii)
<br> Verification of central limit theorem 

For n iid random variables we are going create a new quantity 
$$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_{i}$$
and we are going to study its distribution ( ie histogram) and see if its mimics gaussian properties for large values of n if each $x_{i}$ are distributed according to cauchy

```python
def xbar_array(n,m):
    arr=[ mean(stats.cauchy.rvs(size=n)) for i in range(m)]
    return arr
    
```

we have created a function that creates an array of such $\bar{x}(n)$ containing m such elements.<br>
By default we are going to set m to a large value for smooth plot/histogram generation.
We need to analyse the distribution for n =10 ,50,100,200


```python
xs_10=xbar_array(10,10000)
x10 = linspace(-50,50,1000)
y10 = stats.norm.pdf(x10,mean(xs_10),std(xs_10,ddof=1))
hist(xs_10,bins=100,range=(-50,50),density=True)#n=10
plot(x10,y10)
```

above is the plot when n = 10 and its peaks at x=0 and looks like a cauchy only

```python
xs_50=xbar_array(50,10000)
x50= linspace(-50,50,1000)
y50 = stats.norm.pdf(x50,mean(xs_50),std(xs_50,ddof=1))
hist(xs_50,bins=100,range=(-50,50),density=True) #when n=50
plot(x50,y50)
```

### Question 3: Chi square distribution

We have N IIds who are normally distributed and we are going to study the distribution of the sample variace whose average value is population variance.For N rvs the degree of freedom of that distribution is N-1

```python
#we have to create a function that returns an array of sample variances for sample size n
#for scaling reasons we have to multiply S^2 with n-1
def var_gen(n):
    vs = array([(std(stats.norm.rvs(size=n),ddof=1)**2)*(n-1) for _ in range(10000)])
    return vs
```

Now we need to study the distrbution of sample variances and compare it with $\chi^{2}$ distribution of df = n-1 for n =3,5,10,20

```python
def plot_check(n):
    hist(var_gen(n),density=True,range=(-2,40),bins =90)
    x=linspace(-2,40,1000)
    y=stats.chi2.pdf(x,df=n-1)
    plot(x,y,'k')
    # we have created a function that will simulatneously plot the pdf and histogram
```

```python
figure(figsize=(8,6),dpi=150)
subplot(2,2,1)
plot_check(3)
subplot(2,2,2)
plot_check(5)
subplot(2,2,3)
plot_check(10)
subplot(2,2,4)
plot_check(20)
```

as we can see the histograms and pdfs are a perfect match thus indicating $S^{2}$ aka sample variance or population variance estimator follows a $\chi^{2}$ distribution

### Question 5: Gumbel Distribution

$$\mathbb{P}( \bar{x} - z_{0.5\alpha} \frac{\sigma}{\sqrt{n}} \leq  \mu \leq \bar{x} + z_{0.5\alpha} \frac{\sigma}{\sqrt{n}}) = 1- \alpha$$

Here the interval in which $\mu$ is supposed to lie in for the above probabiliy calculation ie 
$[\bar{x} - z_{0.5\alpha} \frac{\sigma}{\sqrt{n}},\bar{x} - z_{0.5\alpha} \frac{\sigma}{\sqrt{n}}]$ is know as 100 x ($ 1-\alpha $) confidence interval. Here $\bar{x}$ , $\sigma$ and n has usual meanings wrt a sample collection from n IID rv xs who are distributed according to Gumbel.Its impractical to figure out the population standard deviation so we usually take sample variance thus the CLT distribution will be a t distribution instead of a normal one . Now $Z_{\alpha /2 }$ also known as critical value such that $ \mathbb{P}(z > z_{\alpha /2 } ) = \frac{\alpha}{2}$ for z thats is distributed either normally ( for $\sigma$ is known ) or a t function of degree n-1 when we are using sample std. Hence $Z_{\alpha /2} = F(1-\frac{\alpha}{2})$ where F is the corresponding inverse of CDF of the distribution of z

```python
# we are going to create a function that computes and prints the upper and lower bounds
#  of the 95% and 99% confidence interval from the sample size as input.
def CI_bounds(n):
    f=0.95
    xs=array(stats.gumbel_r.rvs(size=n))
    xb=mean(xs)
    sdn=pi/sqrt(6*n)
    zcn=stats.norm.ppf(0.5*(1+f))         
    sdt=std(xs,ddof=1)/sqrt(n)
    zct=stats.t.ppf(0.5*(1+f),df=n-1)
    print("95% CI for normal distribution is [" ,xb - zcn*sdn,",", xb + zcn*sdn,"]")       
    print("95% CI for t distribution is [" ,xb - zct*sdt,",", xb + zct*sdt,"]")         
    f=0.99
    zcn=stats.norm.ppf(0.5*(1+f))         
    zct=stats.t.ppf(0.5*(1+f),df=n-1)
    print("99% CI for normal distribution is [" ,xb - zcn*sdn,",", xb + zcn*sdn,"]")       
    print("99% CI for t distribution is [" ,xb - zct*sdt,",", xb + zct*sdt,"]")                                 
             
             
```

```python
CI_bounds(8)
```

```python
CI_bounds(25)
```

We have found out the confidence intervals for corresponding confidence of 8 rvs and 25 rvs and we can evidently notice that CI for n=25 is narrower than those for n=8 corresponding to the same confidence

### Question 6:Binomial Distribution

##### Part(i)

we are going to compare the stemplot of binomial pmf wrt plot of pdf of gaussian

```python
def compare_plot(p,N):
    xb=[i for i in range(N+1)]
    yb=stats.binom.pmf(xb,N,p)
    stem(xb,yb,linefmt='rx')
    xg=linspace(0,N,1000)
    yg=stats.norm.pdf(xg,N*p,sqrt(N*p*(1-p)))
    plot(xg,yg,'k-')
```

```python
figure(figsize=(20,15))
subplot(3,3,1)
compare_plot(0.25,10)
title("p=0.25 ,N=10 ")
subplot(3,3,2)
compare_plot(0.25,25)
title("p=0.25 ,N=25 ")
subplot(3,3,3)
compare_plot(0.25,100)
title("p=0.25 ,N=100 ")

subplot(3,3,4)
compare_plot(0.5,10)
title("p=0.5 ,N=10 ")
subplot(3,3,5)
compare_plot(0.5,25)
title("p=0.5 ,N=25 ")
subplot(3,3,6)
compare_plot(0.5,100)
title("p=0.5 ,N=100 ")

subplot(3,3,7)
compare_plot(0.75,10)
title("p=0.75 ,N=10 ")
subplot(3,3,8)
compare_plot(0.75,25)
title("p=0.75 ,N=25 ")
subplot(3,3,9)
compare_plot(0.75,100)
title("p=0.75 ,N=100 ")

show()
```

from the 9 plots for different p and N values we see that the gausian is a beautiful approximation for binomial distribution
